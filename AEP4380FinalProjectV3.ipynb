```python
#Zane Blood
#Due December 14th 2022
#AEP 4380 Final Project, Lorenz System
#Processor: 1.4 GHz Quad-Core Intel Core i5
#MacOS Catalina Version 10.15.7
#Compiled in Spyder
#Python version 3.10.8

#NOTE: This script has 4 code sections whose start and end
#are defined by comments. Only one code block should be
#uncommented at any time for the program to run

#The imports
import scipy
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt

#The constants we need
sigma=10
rho=28
beta=8/3

ti=0.0
tf=10.0

r0=np.array([1,2,3])

#The rhs() function
def rhsLorenz(t,r):
    
    #the time derivative vector we return
    drdt=np.zeros([len(r)])
    drdt[0]=sigma*(r[1]-r[0])
    drdt[1]=r[0]*(rho-r[2])-r[1]
    drdt[2]=r[0]*r[1]-beta*r[2]
    
    return drdt

####################################################################
#This section plots the three cross sections at a given final time
#for a constant IC and sgima, rho, beta values
####################################################################

#Apply the solve_ip() to the rhs()
soln=scipy.integrate.solve_ivp(rhsLorenz, [ti, tf], r0, rtol=1.0e-6)

#Plotting
plt.figure(1)
plt.clf()
plt.plot(soln.y[0,:], soln.y[1,:], label='X vs Y', color='c')
plt.plot(np.sqrt(beta*(rho-1)), np.sqrt(beta*(rho-1)), 
         marker='o', color='tab:orange')
plt.plot(-np.sqrt(beta*(rho-1)), -np.sqrt(beta*(rho-1)),
         marker='o', color='tab:orange')
plt.plot(0, 0, marker='o', color='tab:orange')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Cross Section 1")
plt.show()

plt.figure(2)
plt.clf()
plt.plot(soln.y[0,:], soln.y[2,:], label='X vs Z', color='g')
plt.plot(np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
plt.plot(-np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
plt.plot(0, 0, marker='o', color='tab:orange')
plt.xlabel("X")
plt.ylabel("Z")
plt.title("Cross Section 2")
plt.show()

plt.figure(3)
plt.clf()
plt.plot(soln.y[1,:], soln.y[2,:], label='Y vs Z', color='b')
plt.plot(np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
plt.plot(-np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
plt.plot(0, 0, marker='o', color='tab:orange')
plt.xlabel("Y")
plt.ylabel("Z")
plt.title("Cross Section 3")
plt.show()

####################################################################
#END FIRST CODE SECTION
####################################################################



#######################################################################
#This next section animates any of the three cross sections or 3D plots
#As functions of time for a constant IC and sigma, rho, beta values.
#Uncomment the code block to run it
#######################################################################

# #Setup
# numFrames=100

# ti=0.0
# tf=80.0
# dt=(tf-ti)/numFrames

# x=[]
# y=[]
# z=[]

# #Data storing
# for i in range(numFrames):
    
#     #Apply the solve_ip() to the rhs()
#     soln=scipy.integrate.solve_ivp(rhsLorenz, [ti, i*dt], r0, rtol=1.0e-6)
    
#     x.append(soln.y[0,:])
#     y.append(soln.y[1,:])
#     z.append(soln.y[2,:])
    
# %matplotlib

# #Animating / Plotting
# for j in range(numFrames):
    
#     plt.clf()
    
#     #Uncomment which cross section or 3D graph you want to plot:
    
# #     #X vs Y
# #     plt.title("X vs Y, t="+str(j*dt)[0:5])
# #     plt.xlabel("X(t)")
# #     plt.ylabel("Y(t)")
# #    plt.xlim(-30,30)
# #     plt.ylim(-30,30)
# #     plt.plot(x[j], y[j], color='b')
# #     plt.plot(np.sqrt(beta*(rho-1)), np.sqrt(beta*(rho-1)),
# #              marker='o', color='tab:orange')
# #     plt.plot(-np.sqrt(beta*(rho-1)), -np.sqrt(beta*(rho-1)),
# #              marker='o', color='tab:orange')
# #     plt.plot(0, 0, marker='o', color='tab:orange')
    
# #     #X vs Z
# #     plt.title("X vs Z, t="+str(j*dt)[0:5])
# #     plt.xlabel("X(t)")
# #     plt.ylabel("Z(t)")
# #     plt.plot(x[j], z[j],'b')
# #     plt.plot(np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
# #     plt.plot(-np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
# #     plt.plot(0, 0, marker='o', color='tab:orange')
# #     plt.xlim(-20,25)
# #     plt.ylim(0,50)
    
# #     #Y vs Z
# #     plt.title("Y vs Z, t="+str(j*dt)[0:5])
# #     plt.xlabel("Y(t)")
# #     plt.ylabel("Z(t)")
# #     plt.plot(y[j],z[j],'b')
# #     plt.plot(np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
# #     plt.plot(-np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
# #     plt.plot(0, 0, marker='o', color='tab:orange')
# #     plt.xlim(-25,30)
# #     plt.ylim(0,50)

#     #3D plot of the results
#     fig = plt.figure(5)
#     ax = fig.add_subplot(111, projection='3d')

#     ax.plot(x[j], y[j], z[j], 'b', linewidth=1)
#     plt.plot(np.sqrt(beta*(rho-1)), np.sqrt(beta*(rho-1)),
#              rho-1, marker='o', color='tab:orange')
#     plt.plot(-np.sqrt(beta*(rho-1)), -np.sqrt(beta*(rho-1)),
#              rho-1, marker='o', color='tab:orange')
#     plt.plot(0, 0, 0, marker='o', color='tab:orange')
    
# #     ax.set_xlim([-30,30])
# #     ax.set_ylim([-10,50])
# #     ax.set_zlim([-10,50])
    
#     ax.set_xlabel('x(t)')
#     ax.set_ylabel('y(t)')
#     ax.set_zlabel('z(t)')
#     plt.title('Lorenz Attractor, t='+str(j*dt)[0:3])
    
#     plt.show()
#     plt.pause(0.001)

####################################################################
#END SECOND CODE SECTION
####################################################################



####################################################################
#This section animates any of the three cross sections or 3D plots
#As functions of time for 5 different ICs and
#constant sigma, rho, and beta values
####################################################################
    
# r1=np.array([1,2,3])
# r2=np.array([0.5,2,3])
# r3=np.array([1,2.5,3])
# r4=np.array([1,2,3.5])
# r5=np.array([1,1.5,3])

# #Setup
# numFrames=100

# ti=0.0
# tf=80.0
# dt=(tf-ti)/numFrames

# x1=[]
# y1=[]
# z1=[]

# x2=[]
# y2=[]
# z2=[]

# x3=[]
# y3=[]
# z3=[]

# x4=[]
# y4=[]
# z4=[]

# x5=[]
# y5=[]
# z5=[]

# #Data storing
# for i in range(numFrames):
    
#     #Apply the solve_ip() to the rhs() for 5 new IC's (each only off by +or- 0.5)
#     soln1=scipy.integrate.solve_ivp(rhsLorenz, [ti, i*dt], r1, rtol=1.0e-6)
#     soln2=scipy.integrate.solve_ivp(rhsLorenz, [ti, i*dt], r2, rtol=1.0e-6)
#     soln3=scipy.integrate.solve_ivp(rhsLorenz, [ti, i*dt], r3, rtol=1.0e-6)
#     soln4=scipy.integrate.solve_ivp(rhsLorenz, [ti, i*dt], r4, rtol=1.0e-6)
#     soln5=scipy.integrate.solve_ivp(rhsLorenz, [ti, i*dt], r5, rtol=1.0e-6)
    
#     x1.append(soln1.y[0,:])
#     y1.append(soln1.y[1,:])
#     z1.append(soln1.y[2,:])
    
#     x2.append(soln2.y[0,:])
#     y2.append(soln2.y[1,:])
#     z2.append(soln2.y[2,:])
    
#     x3.append(soln3.y[0,:])
#     y3.append(soln3.y[1,:])
#     z3.append(soln3.y[2,:])
    
#     x4.append(soln4.y[0,:])
#     y4.append(soln4.y[1,:])
#     z4.append(soln4.y[2,:])

#     x5.append(soln5.y[0,:])
#     y5.append(soln5.y[1,:])
#     z5.append(soln5.y[2,:])

# %matplotlib

# #Animating / Plotting the different ICs on the same graphs
# for j in range(numFrames):
    
#     plt.clf()
    
#     #Uncomment which cross section or 3D graph you want to plot:
    
#     #X vs Y
#     plt.title("X vs Y, t="+str(j*dt)[0:5])
#     plt.xlabel("X(t)")
#     plt.ylabel("Y(t)")
#     #plt.xlim(-30,30)
#     #plt.ylim(-30,30)
#     plt.plot(x1[j], y1[j], color='r')
#     plt.plot(x2[j], y2[j], color='tab:orange')
#     plt.plot(x3[j], y3[j], color='y')
#     plt.plot(x4[j], y4[j], color='b')
#     plt.plot(x5[j], y5[j], color='g')
#     plt.plot(np.sqrt(beta*(rho-1)), np.sqrt(beta*(rho-1)),
#              marker='o', color='tab:orange')
#     plt.plot(-np.sqrt(beta*(rho-1)), -np.sqrt(beta*(rho-1)),
#              marker='o', color='tab:orange')
#     plt.plot(0, 0, marker='o', color='tab:orange')
    
# #     #X vs Z
# #     plt.title("X vs Z, t="+str(j*dt)[0:5])
# #     plt.xlabel("X(t)")
# #     plt.ylabel("Z(t)")
# #     plt.plot(x1[j], z1[j], color='r')
# #     plt.plot(x2[j], z2[j], color='tab:orange')
# #     plt.plot(x3[j], z3[j], color='y')
# #     plt.plot(x4[j], z4[j], color='b')
# #     plt.plot(x5[j], z5[j], color='g')
# #     plt.plot(np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
# #     plt.plot(-np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
# #     plt.plot(0, 0, marker='o', color='tab:orange')
# #     #plt.xlim(-20,25)
# #     #plt.ylim(0,50)
    
# #     #Y vs Z
# #     plt.title("Y vs Z, t="+str(j*dt)[0:5])
# #     plt.xlabel("Y(t)")
# #     plt.ylabel("Z(t)")
# #     plt.plot(y1[j], z1[j], color='r')
# #     plt.plot(y2[j], z2[j], color='tab:orange')
# #     plt.plot(y3[j], z3[j], color='y')
# #     plt.plot(y4[j], z4[j], color='b')
# #     plt.plot(y5[j], z5[j], color='g')
# #     plt.plot(np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
# #     plt.plot(-np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
# #     plt.plot(0, 0, marker='o', color='tab:orange')
# #     #plt.xlim(-25,30)
# #     #plt.ylim(0,50)

# #     #3D plot of the results
# #     fig = plt.figure(6)
# #     ax = fig.add_subplot(111, projection='3d')

# #     ax.plot(x1[j], y1[j], z1[j], 'r', linewidth=1)
# #     ax.plot(x2[j], y2[j], z2[j], 'tab:orange', linewidth=1)
# #     ax.plot(x3[j], y3[j], z3[j], 'y', linewidth=1)
# #     ax.plot(x4[j], y4[j], z4[j], 'b', linewidth=1)
# #     ax.plot(x5[j], y5[j], z5[j], 'g', linewidth=1)
# #     plt.plot(np.sqrt(beta*(rho-1)), np.sqrt(beta*(rho-1)),
# #              rho-1, marker='o', color='tab:orange')
# #     plt.plot(-np.sqrt(beta*(rho-1)), -np.sqrt(beta*(rho-1)),
# #              rho-1, marker='o', color='tab:orange')
# #     plt.plot(0, 0, 0, marker='o', color='tab:orange')
    
# # #     ax.set_xlim([-30,30])
# # #     ax.set_ylim([-10,50])
# # #     ax.set_zlim([-10,50])
    
# #     ax.set_xlabel('x(t)')
# #     ax.set_ylabel('y(t)')
# #     ax.set_zlabel('z(t)')
# #     plt.title('Lorenz Attractor, t='+str(j*dt)[0:3])
    
#     plt.show()
#     plt.pause(0.001)
    
####################################################################
#END THIRD CODE SECTION
####################################################################



####################################################################
#This section animates the rho dependence for a given start and end
#time and constant IC and sigma and beta values
####################################################################

# %matplotlib

# #Animating the rho dependence from p=11 to p=26 by dp=0.05
# #For a given start postion and cross section and final time
# pi=0.0
# pf=25.0
# dp=0.05

# p=pi

# while p<=pf:
    
#     rho=p
    
#     #Apply the solve_ip() to the rhs()
#     soln=scipy.integrate.solve_ivp(rhsLorenz, [ti, tf], r0, rtol=1.0e-6)
    
#     plt.clf()
    
#     #Uncomment which cross section or 3D graph you want to plot
    
#     #X vs Y
#     plt.title("X vs Y, rho="+str(rho)[0:5])
#     plt.xlabel("X(t)")
#     plt.ylabel("Y(t)")
#     #plt.xlim(-30,30)
#     #plt.ylim(-30,30)
#     plt.plot(soln.y[0,:], soln.y[1,:], label='X vs Y', color='b')
#     plt.plot(np.sqrt(beta*(rho-1)), np.sqrt(beta*(rho-1)),
#              marker='o', color='tab:orange')
#     plt.plot(-np.sqrt(beta*(rho-1)), -np.sqrt(beta*(rho-1)),
#              marker='o', color='tab:orange')
#     plt.plot(0, 0, marker='o', color='tab:orange')
    
# #     #X vs Z
# #     plt.plot(soln.y[0,:], soln.y[2,:], label='X vs Z', color='b')
# #     plt.plot(np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
# #     plt.plot(-np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
# #     plt.plot(0, 0, marker='o', color='tab:orange')
# #     plt.xlabel("X(t)")
# #     plt.ylabel("Z(t)")
# #     plt.title("X vs Z, rho="+str(rho)[0:5])
# # #     plt.xlim(-15, 20)
# # #     plt.ylim(0, 50)
    
# #     #Y vs Z
# #     plt.title("Y vs Z, rho="+str(rho)[0:5])
# #     plt.xlabel("Y(t)")
# #     plt.ylabel("Z(t)")
# #     plt.plot(soln.y[1,:], soln.y[2,:], label='Y vs Z', color='b')
# #     plt.plot(np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
# #     plt.plot(-np.sqrt(beta*(rho-1)), rho-1, marker='o', color='tab:orange')
# #     plt.plot(0, 0, marker='o', color='tab:orange')
# #     #plt.xlim(-25,30)
# #     #plt.ylim(0,50)


# #     #3D plot of the results
# #     fig = plt.figure(5)
# #     ax = fig.add_subplot(111, projection='3d')

# #     ax.plot(soln.y[0,:], soln.y[1,:], soln.y[2,:], 'b', linewidth=1)
# #     plt.plot(np.sqrt(beta*(rho-1)), np.sqrt(beta*(rho-1)),
# #              rho-1, marker='o', color='tab:orange')
# #     plt.plot(-np.sqrt(beta*(rho-1)), -np.sqrt(beta*(rho-1)),
# #              rho-1, marker='o', color='tab:orange')
# #     plt.plot(0, 0, 0, marker='o', color='tab:orange')
    
# # #     ax.set_xlim([-15,20])
# # #     ax.set_ylim([-20,30])
# # #     ax.set_zlim([-0,45])
    
# #     ax.set_xlabel('x(t)')
# #     ax.set_ylabel('y(t)')
# #     ax.set_zlabel('z(t)')
# #     plt.title('Lorenz Attractor, rho='+str(rho)[0:5])
    
    
#     plt.show()
#     plt.pause(0.01)
    
#     p=p+dp
    
# ###################################################################
# END FINAL CODE SECTION
# ###################################################################
```
